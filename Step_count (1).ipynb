```python
#Imports Pandas library for dataframe handling
import pandas as pd

#Imports NumPy library for array handling
import numpy as np

#The glob module finds all the '.xls' pathnames matching to a specific user and stores the data as a dataframe
import glob
glob.glob('*.xls')

#Imports all step_count Excel files for a given subject and appends all 18 months of data for the 'Body' sheet
all_body_data_1 = pd.DataFrame()
for f in glob.glob('2001*'):
            df = pd.ExcelFile(f)
            all_body_data_1 = all_body_data_1.append(df.parse('Body',index_col=0))

#Imports all step_count Excel files for a given subject and appends all 18 months of data for the 'Activities' sheet            
all_activities_data_1 = pd.DataFrame()
for f in glob.glob('2001*'):
            df = pd.ExcelFile(f)
            all_activities_data_1 = all_activities_data_1.append(df.parse('Activities',index_col=0))

#Concatenates all 18 months of subject data for 'Body' and 'Activities' sheets             
p2001=pd.concat([all_body_data_1,all_activities_data_1],axis=1)

#Adds patient ID column
p2001['studyID']=2001

p2001
```
